```python
def hyperbolic_paraboloid(x, a=1, b=1):
    if type(x) == np.ndarray:
        if x.shape[1] == 2:
            x = x.T
    return (x[0]/a) ** 2 - (x[1]/b) ** 2
```

```python
def visualize_fw(obj_func):
    xcoord = np.linspace(-10.0,10.0,50)
    ycoord = np.linspace(-10.0,10.0,50)
    w1,w2 = np.meshgrid(xcoord,ycoord)
    pts = np.vstack((w1.flatten(),w2.flatten()))
    
    # All 2D points on the grid
    pts = pts.transpose()
    
    # Function value at each point
    f_vals = obj_func(pts)
    function_plot(pts,f_vals)
    plt.title('Objective Function Shown in Color')
    plt.show()
    return pts,f_vals

# Helper function to annotate a single point
def annotate_pt(text,xy,xytext,color):
    plt.plot(xy[0],xy[1],marker='P',markersize=10,c=color)
    plt.annotate(text,xy=xy,xytext=xytext,
                 # color=color,
                 arrowprops=dict(arrowstyle="->",
                 color = color,
                 connectionstyle='arc3'))

# Plot the function
# Pts are 2D points and f_val is the corresponding function value
def function_plot(pts, f_val):
    f_plot = plt.scatter(pts[:,0],pts[:,1],
                         c=f_val,vmin=min(f_val),vmax=max(f_val),
                         cmap='RdBu_r')
    plt.colorbar(f_plot)
    # Show the optimal point
    annotate_pt('Inflection point',(0,0),(-5,-7),'yellow')    

pts,f_vals = visualize_fw(hyperbolic_paraboloid)
```

```python
# Objective function
def hyperbolic_paraboloid(x, a=1, b=1):
    if type(x) == np.ndarray:
        if len(x.shape) == 1:
            return (x[0]/a) ** 2 - (x[1]/b) ** 2
        if x.shape[1] == 2:
            x = x.T
    return (x[0]/a) ** 2 - (x[1]/b) ** 2

# Function to compute the gradient
def grad(x, a=1, b=1):
    if type(x) == np.ndarray:
        if len(x.shape) == 1:
            return np.array([2*(x[0]/a),-2*(x[1]/b)])
        if x.shape[1] == 2:
            x = x.T
            return np.array([2*(x[0]/a), -2*(x[1]/b)]).T
    return np.array([2*(x[0]/a), -2*(x[1]/b)])


# Make threshold a -ve value if you want to run exactly
# max_iterations.
def gradient_descent(max_iterations, threshold, w_init,
                     obj_func, grad_func,
                     learning_rate=0.05, momentum=0.8,optimization=False, noize=None):
    
    w = w_init
    w_history = w
    f_history = obj_func(w)
    delta_w = np.zeros(w.shape)
    i = 0
    diff = 1.0e10
    
    while  i<max_iterations and diff>threshold:
        delta_w = -learning_rate*grad_func(w) + momentum*delta_w
        w = w + delta_w
        if noize is not None:
            w = w + noize*(w+w.max())*np.random.normal(size=len(w_init))/100
        
        # store the history of w and f
        w_history = np.vstack((w_history,w))
        f_history = np.vstack((f_history,obj_func(w)))
        
        # update iteration number and diff between successive values
        # of objective function
        i+=1
        diff = np.absolute(f_history[-1]-f_history[-2])
    if optimization == True:
        return w,obj_func(w),i
    return w_history,f_history

# Function to plot the objective function
# and learning history annotated by arrows
# to show how learning proceeded
def visualize_learning(w_history):  
    
    # Make the function plot
    function_plot(pts,f_vals)
    
    # Plot the history
    plt.plot(w_history[:,0],w_history[:,1],marker='o',c='magenta') 
    
    # Annotate the point found at last iteration
    annotate_pt('end point',
                (w_history[-1,0],w_history[-1,1]),
                (-1,7),'green')
    iter = w_history.shape[0]
    for w,i in zip(w_history,range(iter-1)):
        # Annotate with arrows to show history
        plt.annotate("",
                    xy=w, xycoords='data',
                    xytext=w_history[i+1,:], textcoords='data',
                    arrowprops=dict(arrowstyle='<-',
                            connectionstyle='angle3'))     
    
def solve_fw(w_init, learning_rates = [0.01, 0.03, 0.05, 0.1],max_iterations=5, noize=None):
    # Setting up

    fig, ax = plt.subplots(nrows=4, ncols=4, figsize=(18, 12))
    momentum = [0, 0.5, 0.9]
    ind = 1
    
    # Iteration through all possible parameter combinations
    for alpha in momentum:
        for eta,col in zip(learning_rates,[0,1,2,3]):
            plt.subplot(3,4,ind)        
            w_history,f_history = gradient_descent(max_iterations,-1,w_init, hyperbolic_paraboloid,grad, eta, alpha, noize=noize)
            
            visualize_learning(w_history)
            ind = ind+1
            plt.text(-9, 12,'Learning Rate = '+str(eta),fontsize=13)
            if col==1:
                plt.text(10,15,'momentum = ' + str(alpha),fontsize=20)

    fig.subplots_adjust(hspace=0.5, wspace=.3)
    plt.show()
```

```python
for i in range(3,5):
    rand = np.random.RandomState(i)
    w_init = rand.uniform(-8,8,2)
    w_init[1] = w_init[1]/4 # для того чтобы не начинать сразу "на краю"
    print(w_init)
    solve_fw(w_init)
    print('\n\n\n\n')
w_init = np.array([7, 0])
print(w_init)
solve_fw(w_init, learning_rates=[0.05, 0.2, 0.5, 0.8])
    
```

## 2. Рассмотреть несколько случаев для начальных приближений с разной удаленностью от искомых точек.


Т.к. у функции нет ни локальных ни глобальных екстремумов - то в общем случае алгоритм никуда не сходится - это видно из первых двух примеров визуализции, единственным исключением является ситуация, когда y=0:\
Т.к градиент в этом случае = 0 по y, то мы будем оставаться на этой плоскости, следовательно это единственный случай, подходящий под рассмотрение

**Подбор оптимальных параметров**

```python
min_iter = 150
best_params = {}
learning_rates = [0.01, 0.03, 0.05, 0.1]
momentums = [0, 0.5, 0.9]
for lr in learning_rates:
    for momentum in momentums:
        optim, minimum, n_iter = \
        gradient_descent(
            max_iterations=100,
            threshold= 1e-10,
            w_init=np.array([7, 0]),
            obj_func=hyperbolic_paraboloid,
            grad_func=grad,
            learning_rate=lr,
            momentum=momentum,
            optimization=True
    )
        if n_iter < min_iter:
            min_iter = n_iter
            best_params['lr'] = lr
            best_params['momentum'] = momentum
```

```python
best_params
```

```python
for p in np.array([[100, 0],[10, 0], [1, 0], [7, 0]]):
    optim, minimum, n_iter = gradient_descent(
            max_iterations=100,
            threshold= 1e-10,
            w_init=p,
            obj_func=hyperbolic_paraboloid,
            grad_func=grad,
            learning_rate=best_params['lr'],
            momentum=best_params['momentum'],
            optimization=True
    )
    print(f"Minimum = {minimum} at {optim}, iterations = {n_iter}")
```

Видим, что градиентный спуск сошёлся везде не более чем за 45 итераций

```python
from scipy.optimize import minimize
for p in np.array([[1000000, 0],[10, 0], [1, 0], [7, 0]]):
    result = minimize(hyperbolic_paraboloid, p, method='SLSQP', tol=1e-10, constraints={'type': 'eq', 'fun': lambda x:  x[1]})
    print(f"Minimum g(x,y) = {result.fun} found at {result.x} after {result.nit} iterations")
```

```python
from scipy.optimize import minimize
for p in np.array([[1000000, 0],[10, 0], [1, 0], [7, 0]]):
    result = minimize(hyperbolic_paraboloid, p, method='trust-constr', tol=1e-10, constraints={'type': 'eq', 'fun': lambda x:  x[1]})
    print(f"Minimum g(x,y) = {result.fun} found at {result.x} after {result.nit} iterations")
```

С ограничениями (y=0) встренный метод SLSQP работает значительно быстрей: даже при существенных отклонениях быстро достигает "оптимума".
метод trust-constr работает примерно схоже с "ручным" градиентным спуском.
