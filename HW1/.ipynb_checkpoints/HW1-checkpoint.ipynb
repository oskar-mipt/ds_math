### 1. Даны спектры поглощения 6-ти сложных органических вещества в формате spc.

**Из-за ошибок в модуле pyspectra не удаётся прочитать файлы spc, загружаю их преобразовав в json в другом источнике**

```python
path = './spectras-20220122/spectras/'
spectras_files = [f for f in os.listdir(path) if os.path.isfile(os.path.join(path, f))]
```

```python
from pyspectra.readers.read_spc import read_spc
spectras = {}
for n,i in enumerate(spectras_files):
    spectras[n] = read_spc(path+i)
spectras = {i:j.to_dict() for i,j in spectras.items()}
```

```python
import json

with open('spectras.json') as f:
    spectras = json.load(f)
spectras = {int(i):pd.Series(j.values(),index=[float(i) for i in j.keys()]) for i,j in spectras.items()}
```

```python
df = pd.concat([
    i for i in spectras.values()
],axis=1)
df[(df[0].index>493.68)& (df[0].index<493.76)]
```

**Видим, что замеры концентраций были в разное время, значит нужно понять, как можно привести это к полному датасету**

```python
for i in range(len(spectras)):
    print(i, (spectras[i].index[1:] - spectras[i].index[:-1]).min())
```

**Видим, что шаги замеров для всех веществ примерно одинаковы, значит, надо примерно с этим шагом их "склеить"**

**"Приклеивать" будем к измерениям первого вещества**

```python
step = 0.0622
```

```python
new_spectras = []
for i in tqdm_notebook(spectras[0].index):
    new_spectras.append([i]+
        [next(iter(list(spectras[sp_n][(spectras[sp_n].index>=i)&(spectras[sp_n].index<i+step)]) or []), np.nan) 
             for sp_n in range(len(spectras))]
    )
```

```python
df = pd.DataFrame(new_spectras,columns=['index',0,1,2,3,4,5])
df = df.set_index('index')
df = df[df.notnull().min(axis=1)]
```

```python
plt.figure(dpi=150)
plt.plot(df)
plt.legend()
plt.show()
```

### 2. Необходимо составить линейную комбинацию смеси 4 данных веществ с рандомными концентрациями, а затем, решая переопределенную систему линейных алгебраических уравнений, определить концентрации данных веществ, посчитать невязку по отношению к известным (исходным) концентарциям

```python
import random

random_state = 2
```

```python
# Выбираем рандомные 4 смеси
random.seed(random_state)
substance_ids = random.sample(range(len(spectras)),k=4,)
print(substance_ids)
```

```python
# Выбираем рандомные концентрации
np.random.seed(random_state)
concentrations = np.random.random(size = 4)
concentrations = concentrations/concentrations.sum()
print(concentrations)
```

```python
# Формируем смесь
df['mix'] = (df.iloc[:,substance_ids]*concentrations).sum(axis=1).values
```

**Решаем обратную задачу: находим коэффициенты, используем линейную регрессию**

```python
import statsmodels.api as sm

res = sm.OLS(endog=df[['mix']].values, exog=df.iloc[:,substance_ids].values).fit()
# res = mod.fit()
```

```python
res.params
```

```python
concentrations
```

```python
print('Sums of squared residuals:',((res.params-concentrations)**2).sum())
print('Sums of abs residuals:',(abs(res.params-concentrations)).sum())
```

**Также можно решать систему (МНК)**

```python
params = np.linalg.lstsq(df.iloc[:,substance_ids].values, df[['mix']].values,rcond=None)
```

```python
print(params[0].T[0])
```

```python
print('Sums of squared residuals:',((params[0].T[0]-concentrations)**2).sum())
print('Sums of squared residuals:',(abs(params[0].T[0]-concentrations)).sum())
```

### 3.Найти локальные максимумы и минимумы спектра поглощения в заданном диапазоне длин волн с субпиксельной точностью.

```python
from scipy.interpolate import CubicSpline
```

```python
#Заданный диапазон
int_range = [1500,2600]
```

```python
# точки для интерполяции
x = np.linspace(int_range[0], int_range[1], 100)
x_ = np.linspace(int_range[0], int_range[1], 1000)
```

```python
interp_f = np.interp(x, df['mix'].index, df['mix'].values,)
interp_f2 = CubicSpline(x, interp_f, bc_type='not-a-knot')
# 
```

```python
plt.figure(dpi=350)
plt.plot(df.loc[(df.index>=int_range[0])&(df.index<int_range[1]),'mix'])
plt.plot(x_[1:-1], interp_f2(x_[1:-1]),alpha=1,color='orange')
```

**Локальные максимумы и минимумы**

```python
from scipy.signal import argrelextrema

loc_maxs = argrelextrema(interp_f2(x_[1:-1]), np.greater)
loc_mins = argrelextrema(interp_f2(x_[1:-1]), np.less)

plt.figure(dpi=350)
plt.plot(x_[1:-1], interp_f2(x_[1:-1]),alpha=1,color='orange',zorder=1)
plt.scatter(x_[1:-1][loc_maxs],interp_f2(x_[1:-1])[loc_maxs],s=5,label='max',color='red')
plt.scatter(x_[1:-1][loc_mins],interp_f2(x_[1:-1])[loc_mins],s=5,label='min',color='blue')
plt.legend()
plt.show()
```
